# Phase1: Camera calibration

## Execute camera calibration looking for chessboard corners

```python
import numpy as np
import cv2
import glob
import pickle
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

nx = 9 # num corners on chessboard horizontal axis
ny = 6 # num corners on chessboard vertical axis

# cv2.cornerSubPix termination criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(nx,ny,0)
objp = np.zeros((ny*nx,3), np.float32)
objp[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d point in real world space
imgpoints = [] # 2d points in image plane.

images = glob.glob('camera_cal/*.jpg')
print('Found {} images\n'.format(len(images)))
img_counter = 0

for fname in images:
    img_counter += 1
    print('Processing image {}'.format(img_counter))
    
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Find the chess board corners
    ret, corners = cv2.findChessboardCorners(gray, (nx,ny),None)


    # If found, add object points and image points
    if ret == True:
        objpoints.append(objp)

        corners2 = cv2.cornerSubPix(gray,corners,(11,11),(-1,-1),criteria)
        imgpoints.append(corners2)

        # Draw and display the corners
        img = cv2.drawChessboardCorners(img, (nx,ny), corners2, ret)
        #cv2.imshow('img',img)
        #cv2.waitKey(500)
        
        #Example of corners finding on calibration3 image
        if fname=='camera_cal/calibration2.jpg':
            f, (ax1, ax2) = plt.subplots(2, 1, figsize=(30, 15))
            f.tight_layout()
            ax1.imshow(mpimg.imread(fname))
            ax1.set_title('Original Image', fontsize=12)
            ax2.imshow(img)
            ax2.set_title('Processed Image', fontsize=12)
            plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)

#cv2.destroyAllWindows()

print('Saving points data...')
wide_dist_pickle = {}
wide_dist_pickle['objpoints'] = objpoints
wide_dist_pickle['imgpoints'] = imgpoints

pickle_out = open('wide_dist_pickle.pkl','wb')
pickle.dump(wide_dist_pickle, pickle_out)
pickle_out.close()
```

## Generate undistort image using camera calibration data

```python
# Read in the saved objpoints and imgpoints
print('Loading points data...')
wide_dist_pickle = pickle.load( open( 'wide_dist_pickle.pkl', 'rb' ) )
objpoints = wide_dist_pickle["objpoints"]
imgpoints = wide_dist_pickle["imgpoints"]

# Read in an image from camera_cal image set
img = cv2.imread('camera_cal/calibration2.jpg')

# Write a function that takes an image, object points, and image points
# performs the camera calibration, image distortion correction and 
# returns the undistorted image
def cal_undistort(img, objpoints, imgpoints):
    # Use cv2.calibrateCamera() and cv2.undistort()
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    ret,mtx,dist,rvec,tvec = cv2.calibrateCamera(objpoints,imgpoints,gray.shape[::-1],None,None)
    undist = cv2.undistort(img,mtx,dist,None,mtx)
    return undist,mtx,dist

undistorted,mtx,dist = cal_undistort(img, objpoints, imgpoints)

# Save distorsion coefficients
print('Saving distorsion coefficients...')
wide_dist_pickle['mtx'] = mtx
wide_dist_pickle['dist'] = dist

pickle_out = open('wide_dist_pickle.pkl','wb')
pickle.dump(wide_dist_pickle, pickle_out)
pickle_out.close()

# Show result
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
f.tight_layout()
ax1.imshow(img)
ax1.set_title('Original Image', fontsize=12)
ax2.imshow(undistorted)
ax2.set_title('Undistorted Image', fontsize=12)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

# Step2: Thresholding

## HLS and Color Threshold

```python
orig_img = mpimg.imread('test_images/straight_lines1.jpg')

def binary_threshold_pipeline(img, s_thresh=(150, 255), sx_thresh=(20, 100)):
    img = np.copy(img)
    # Convert to HLS color space and separate the V channel
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    l_channel = hls[:,:,1]
    s_channel = hls[:,:,2]
    # Sobel x
    sobelx = cv2.Sobel(l_channel, cv2.CV_64F, 1, 0) # Take the derivative in x
    abs_sobelx = np.absolute(sobelx) # Absolute x derivative to accentuate lines away from horizontal
    scaled_sobel = np.uint8(255*abs_sobelx/np.max(abs_sobelx))
    
    # Threshold x gradient
    sxbinary = np.zeros_like(scaled_sobel)
    sxbinary[(scaled_sobel >= sx_thresh[0]) & (scaled_sobel <= sx_thresh[1])] = 1
    
    # Threshold color channel
    s_binary = np.zeros_like(s_channel)
    s_binary[(s_channel >= s_thresh[0]) & (s_channel <= s_thresh[1])] = 1
    # Stack each channel
    color_binary = np.dstack(( np.zeros_like(sxbinary), sxbinary, s_binary)) * 255
    
    # Convert to binary image
    thresh = (0, 255)
    gray = cv2.cvtColor(color_binary, cv2.COLOR_RGB2GRAY)
    binary = np.zeros_like(gray)
    binary[(gray > thresh[0]) & (gray <= thresh[1])] = 1
    
    return binary
    
binary_thres_img = binary_threshold_pipeline(orig_img)


# Show result
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
f.tight_layout()
ax1.imshow(orig_img)
ax1.set_title('Original Image', fontsize=12)
ax2.imshow(binary_thres_img, cmap='gray')
ax2.set_title('Binary Image', fontsize=12)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

# Step 3: Undistort and transform

## Undistort and transform image

```python
def getPerspectiveParams():
    # define 4 source points points_src = np.float32([[,],[,],[,],[,]])
    pA = np.array([180,719])
    pB = np.array([598,447])
    pC = np.array([680,447])
    pD = np.array([1130,719])
    points_src = np.float32([pA,pB,pC,pD])    
    # define 4 destination points points_dst = np.float32([[,],[,],[,],[,]])
    pA1 = np.array([300,719])
    pB1 = np.array([300,0])
    pC1 = np.array([900,0])
    pD1 = np.array([900,719])
    points_dst = np.float32([pA1,pB1,pC1,pD1])     
    # use cv2.getPerspectiveTransform() to get M, the transform matrix
    M = cv2.getPerspectiveTransform(points_src,points_dst)
    Minv = cv2.getPerspectiveTransform(points_dst, points_src)
    return M, Minv


def undist_and_transform(img,mtx,dist,M):
    img_size = (img.shape[1], img.shape[0])
    undist_img = cv2.undistort(img,mtx,dist,None,mtx)    
    # use cv2.warpPerspective() to warp your image to a top-down view
    warped = cv2.warpPerspective(undist_img, M, img_size, flags=cv2.INTER_LINEAR)
    return undist_img, warped


M, Minv = getPerspectiveParams()

# Save perspective parameters
print('Saving perspective parameters...')
wide_dist_pickle['M'] = M
wide_dist_pickle['Minv'] = Minv

pickle_out = open('wide_dist_pickle.pkl','wb')
pickle.dump(wide_dist_pickle, pickle_out)
pickle_out.close()

undist_img, warped = undist_and_transform(orig_img,mtx,dist,M)

# Show result
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
f.tight_layout()
ax1.imshow(undist_img)
ax1.set_title('Undistorted Image', fontsize=12)
ax2.imshow(warped)
ax2.set_title('Undistorted and Warped Image', fontsize=12)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)

binary_undist_img, binary_warped = undist_and_transform(binary_thres_img,mtx,dist,M)

# Show result
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))
f.tight_layout()
ax1.imshow(binary_undist_img, cmap='gray')
ax1.set_title('Binary Thresholded Image', fontsize=12)
ax2.imshow(binary_warped, cmap='gray')
ax2.set_title('Binary Undistorted and Warped Image', fontsize=12)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

# Finding lane lines

```python
def hist(input_img):
    # Grab only the bottom half of the image
    # Lane lines are likely to be mostly vertical nearest to the car
    img = input_img.copy()
    half_index = img.shape[0]//2
    # clean the borders
    img[half_index:,:50] = 0
    img[half_index:,-50:] = 0
    bottom_half = img[half_index:,:]

    # Sum across image pixels vertically - make sure to set `axis`
    # i.e. the highest areas of vertical lines should be larger values
    histogram = np.sum(bottom_half,axis=0)
    
    return histogram

# Create histogram of image binary activations
histo = hist(binary_warped)

# Visualize the resulting histogram
plt.plot(histo)
```

```python
def find_lane_pixels_sliding_windows(binary_warped, left_lane_line, right_lane_line):
    # Create histogram of image binary activations
    histogram = hist(binary_warped)   
    
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])    
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint
    
    left_lane_line.set_xline_base_pixels(leftx_base)
    right_lane_line.set_xline_base_pixels(rightx_base)
            
    # HYPERPARAMETERS
    # Choose the number of sliding windows
    nwindows = 9
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 50
    
    # Set height of windows - based on nwindows above and image shape
    window_height = np.int(binary_warped.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero (i.e. activated) pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base
    
    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []
    
    '''
    Now that we've set up how the windows look like and have a starting point, 
    we want to loop for nwindows, with the given window sliding left or right 
    if it finds the mean position of activated pixels within the window to have 
    shifted.
    '''
    
    # Step through the windows one by one
    for window in range(nwindows):        
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height # pixel identfies the upper border
        win_y_high = binary_warped.shape[0] - window*window_height # pixel identfies the lower border
        ### Find the four below boundaries of the window ###
        win_xleft_low = leftx_current - margin  # Update this
        win_xleft_high = leftx_current + margin  # Update this
        win_xright_low = rightx_current - margin  # Update this
        win_xright_high = rightx_current + margin  # Update this
    
        ### Identify the nonzero pixels in x and y within the window ###
        true_map_xleft = (nonzerox >= win_xleft_low) & (nonzerox < win_xleft_high)
        true_map_xright = (nonzerox >= win_xright_low) & (nonzerox < win_xright_high)
        true_map_y = (nonzeroy >= win_y_low) & (nonzeroy < win_y_high)
        
        true_map_leftwin = true_map_xleft & true_map_y
        true_map_rightwin = true_map_xright & true_map_y    
        
        good_left_inds = np.where(true_map_leftwin)[0]        
        good_right_inds = np.where(true_map_rightwin)[0]
                
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        ### If you found > minpix pixels, recenter next window ###
        ### (`right` or `leftx_current`) on their mean position ###
        if len(good_left_inds) > minpix:        
            # prima recupero il valore delle colonne dove i pixel sono nonzero 
            # sulla finestra di sinistra
            good_left_cols = nonzerox[good_left_inds]
            # poi uso il valore delle colonne dei pixel per ricentrare 
            # la finestra di sinistra
            leftx_current = np.int(np.mean(good_left_cols))
        if len(good_right_inds) > minpix:        
            good_right_cols = nonzerox[good_right_inds]
            rightx_current = np.int(np.mean(good_right_cols))
    
    # Concatenate the arrays of indices (previously was a list of lists of pixels)
    try:
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
    except ValueError:
        # Avoids an error if the above is not implemented fully
        pass
    
    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
    
    # Save pixel positions inside lane_lines
    left_lane_line.set_all_pix_values(leftx, lefty)
    right_lane_line.set_all_pix_values(rightx, righty)
```

```python
def fit_polynomial(left_lane_line, right_lane_line):    
    
    # Generate y values for plotting
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0])
    
    ### Fit a second order polynomial to each using `np.polyfit` ###
    if left_lane_line.is_detected():
        leftx, lefty = left_lane_line.get_all_pix_values()        
        left_fit = np.polyfit(lefty,leftx,2)
        left_lane_line.set_curr_poly_coeffs(left_fit)
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        left_lane_line.set_current_xfitted(left_fitx)
        
    if right_lane_line.is_detected():
        rightx, righty = right_lane_line.get_all_pix_values()
        right_fit = np.polyfit(righty,rightx,2)    
        right_lane_line.set_curr_poly_coeffs(right_fit)
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
        right_lane_line.set_current_xfitted(right_fitx)
```

```python
def poly_visual(binary_warped,left_lane_line,right_lane_line):
    
    # Create an output image to draw on and visualize the result
    out_img = np.dstack((binary_warped, binary_warped, binary_warped))
    
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0])
    
    if left_lane_line.is_detected():
        print('left line detected')
        leftx,lefty = left_lane_line.get_all_pix_values()
        left_fitx = left_lane_line.get_current_xfitted()
        out_img[lefty, leftx] = left_lane_line.get_lane_pixels_color()
        plt.plot(left_fitx, ploty, color='yellow')
    
    if right_lane_line.is_detected():
        print('right line detected')
        rightx,righty = right_lane_line.get_all_pix_values()
        right_fitx = right_lane_line.get_current_xfitted()  
        out_img[righty, rightx] = right_lane_line.get_lane_pixels_color()
        plt.plot(right_fitx, ploty, color='yellow')
        
    return out_img
```

## Unwarp process

```python
def unwarp(undist_img, warped, left_lane_line, right_lane_line, lane_color=(0,255, 0)):
    result = undist_img.copy()
    # Create an image to draw the lines on
    warp_zero = np.zeros_like(warped).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))

    print(color_warp.shape)
    
    ploty = np.linspace(0, warped.shape[0]-1, warped.shape[0])
    
    if left_lane_line.is_detected() and right_lane_line.is_detected():    
        # Recast the x and y points into usable format for cv2.fillPoly()
        left_fitx = left_lane_line.get_current_xfitted()
        #pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])

        right_fitx = right_lane_line.get_current_xfitted()
        #pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
        
        px = np.concatenate([left_fitx,np.flipud(right_fitx)])
        py = np.concatenate([ploty,np.flipud(ploty)])

        pts = np.array( [[x,y] for x,y in zip(px,py)], np.int32)        
        
        # Draw the lane onto the warped blank image
        cv2.fillPoly(color_warp, [pts], lane_color)

        # Warp the blank back to original image space using inverse perspective matrix (Minv)
        img_size = (undist_img.shape[1], undist_img.shape[0])
        newwarp = cv2.warpPerspective(color_warp, Minv, img_size ,flags=cv2.INTER_LINEAR) 
        # Combine the result with the original image
        result = cv2.addWeighted(undist_img, 1, newwarp, 0.3, 0)
    return result
```

```python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) 
    # for processing video below put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)
    
    undist_img,_ = undist_and_transform(image,mtx,dist,M)
    
    binary_thres_img = binary_threshold_pipeline(image)
    binary_undist_img, binary_warped = undist_and_transform(binary_thres_img,mtx,dist,M)
    
    find_lane_pixels_sliding_windows(binary_warped,left_lane_line,right_lane_line)
    fit_polynomial(left_lane_line,right_lane_line)

    result = unwarp(undist_img, binary_warped, left_lane_line, right_lane_line)
    
    return result
```
